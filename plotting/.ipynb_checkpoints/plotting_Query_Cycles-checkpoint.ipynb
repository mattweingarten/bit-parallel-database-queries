```python
import pandas as pd

data_q = pd.read_csv('../runtimedata/q1_parallel_weave.csv',header=None,names=['cycles','rows','cols','gen']) 
data_qb = pd.read_csv('../runtimedata/q1_weave.csv',header=None,names=['cycles','rows','cols','gen']) 
#data_q = data_q.apply(lamda a: str(a) + 'parallel')

cols = data_q['cols'].unique()
rows = data_q['rows'].unique()
gen = data_q['gen'].unique()

data_gg = data_q[data_q.gen == 1]
data_ggb = data_qb[data_qb.gen == 1]
data_gg['cols'] = data_gg['cols'].apply((lambda a: 'parallel ' + str(a)))
data_ggb['cols'] = data_ggb['cols'].apply((lambda a: str(a)))

data_gg = data_gg.append(data_ggb)
data_cols = data_gg.pivot(index='rows',values='cycles',columns='cols')

plt = data_cols.plot(ylim=(0,3e7),figsize=(15,10))

```

```python
data_colsb = data_ggb.pivot(index='rows',values='cycles',columns='cols')
data_colsb.plot()
```
